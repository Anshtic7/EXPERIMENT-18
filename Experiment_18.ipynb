```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
np.random.seed(10)
df = pd.DataFrame({
    'Category': ['A','B','C','D','E'],
    'Values': [0,70,30,90,60],
    'Sales': np.random.randint(100,500,5),
    'Profit': np.random.randint(10,100,5)
})
print(df.head())
```

# **Area Plot**

```python
plt.figure()
plt.fill_between(df['Category'],df['Values'],color='green',alpha=0.5)
plt.title('Ansh')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

**2 LINE GRAPHS**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

plt.figure(figsize=(7, 4))

plt.fill_between(df['Category'], df['Values'], color='green', alpha=0.5, label='Values')
plt.fill_between(df['Category'], df['Sales'], color='yellow', alpha=0.5, label='Sales')

plt.title('Ansh')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.show()
```

# **PIE CHART**

```python
plt.figure()
plt.pie(df['Values'],labels=df['Category'],autopct='%1.2f%%')
plt.title('Ansh')
plt.show()
```

# **DONUT CHART**

```python
plt.figure()
plt.pie(df['Values'],labels=df['Category'],autopct='%1.1f%%')
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Ansh')
plt.show()
```

# **BOXPLOT**

```python
plt.figure()
sns.boxplot(x=df['Values'])
plt.title('Ansh')
plt.show()
```

# **HEATMAP**

```python
plt.figure()
corr = df[['Sales','Profit']].corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title('Ansh')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df[['Sales', 'Profit', 'Values']].corr()
plt.figure(figsize=(5, 4))
sns.heatmap(corr, annot=True)
plt.title('Ansh')
plt.show()
```

# **BUBBLE PLOT**

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df['Sales'], df['Profit'], s=df['Values']*10, color = 'red', alpha=0.5)
plt.title('Ansh')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.scatterplot(x='Sales', y='Profit', data=df, size='Values', hue = 'Values', sizes=(50, 300), palette = 'viridis', alpha=0.5)
plt.title('Ansh')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```
